```python
#importing the pymongo
import pymongo
from bson.objectid import ObjectId
```

```python
#Then we will need a specific client to access the database
from pymongo import MongoClient
#The client will need the node_id:port of the bluehive node on whilch mongodb is running in background
client = MongoClient('bhc0176',27017)

```

```python
#Then we need to access a particular database. To see the list of databases present in this database, we can do the following:
client.list_database_names()
#THe client api is available here: http://api.mongodb.com/python/current/api/pymongo/mongo_client.html
```

```python
#Then to access a specific database, we can do the following
db = client["last_ditch_effort"]
#All the database related api ai available here:http://api.mongodb.com/python/current/api/pymongo/database.html#pymongo.database.Database
```

```python
#A database is made of made of several Collections which are roughly equivalent to relational tables. To see the list of tables associated
#with a database, we can do:
db.list_collection_names()
#All the Collection related info can be found here: http://api.mongodb.com/python/current/api/pymongo/collection.html#pymongo.collection.Collection
```

```python
#To access a particular collection within the database, we can do:
metrics_collection = db.metrics
#To run a very simple query o this collection, we can use collection_name.find(). It returns a cursos object on which we can traverse by
#To find all about cursor,visist http://api.mongodb.com/python/current/api/pymongo/cursor.html#pymongo.cursor.Cursor
metrics_all = metrics_collection.find()

#To count the number of documents retrieved by that query, we can do:
num_match = metrics_all.count()
#As this one is deprecated, we can use 
num_match = metrics_collection.count_documents({})
#And in the {}, we can give any query we like
print(num_match)
```

```python
#Now, say we want to get the best results derived from the model, We can simply sort the query based on our desired
#parameter and return the result
#To load it in jupyter:
import pandas as pd
db = client.gpu_multi_trans_same_hypers_as_orig
#We will search for the best result. For that, we need to give where the best result is:
result_field = "info.final_result.accuracy"
#We will sort them descendingly. SO, we are using -1
#How many results we want to see in in num_results_selected
num_results_selected = 10
query_res = db.runs.find().sort([(result_field,-1)]).limit(num_results_selected)
#t([("number", 1), ("date", -1)])
#[("number", 1), ("date", -1)]
#.sort({"info.final_result.accuracy":-1})

df = pd.DataFrame(list(query_res)).iloc[0]

#df = pd.DataFrame(list(query_res)).iloc[0]["info"]
#.find({"date": {"$lt": d}}).sort("author"):
info = df["info"]
info
#query_res[0]
#df
```
